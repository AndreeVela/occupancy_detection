```python
import pandas as pd
```

```python
def test( x ):
    if x < 0: return '-'
    if x == 0: return '0'
    
    return '+'

df = pd.read_csv( 'testing_dataset.csv' )
```

```python
df[ 'score_change' ] = df[ 'score_change' ].transform( test )
df
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, RepeatedStratifiedKFold, cross_val_score

cv_method = RepeatedStratifiedKFold( n_splits = 5, n_repeats = 1, random_state = 0 )

estimator = KNeighborsClassifier( n_neighbors = 3 )

cross_val_score( 
    estimator, 
    df.drop( columns = [ 'score_change' ] ), 
    df.score_change, 
    cv = cv_method, 
    scoring = 'accuracy' ).mean()
```
